> *5.4.1.1 Platelet Count*

>> Emphysema is often accompanied by lung tissue damage, often due to smoking, oxidative stress, or environmental pollutants. This leads to pro-inflammatory cytokines such as interleukin-6 (IL-6), tumor necrosis factor-alpha (TNF-), and interleukin-1 (IL-1) being released. These cytokines stimulate megakaryocytes in the bone marrow to increase platelet production to respond to the injured alveolar tissue. As such, increased platelet count in the lungs correlates to emphysema, and has been associated with higher disease exacerbations.

>*5.4.1.2 Blood Eosinophils*

>> Around 20-40% of COPD patients, inclusive of emphysema patients, experience T2 inflammation, where allergens or other triggers activate Th2 immune cells, which release cytokines such as IL-13, IL-5. IL-5 in particular, stimulates the growth and production of eosinophils, which are immune cells, in the bone marrow. Eosinophils then contribute to airway inflammation by releasing cytokines and chemokines to fight off irritants in the lungs, however they also damage lung tissue and increase mucus secretion. Studies show that elevated levels of eosinophils indicate severe exacerbations of emphysema, and that patients with T2 inflammation respond better to inhaled corticosteroids (ICS) treatment for emphysema.


>*5.4.1.3 Segmented Neutrophils Percent*

>> Segmented neutrophils are recruited from the bloodstream to the lungs to fight off irritants such as cigarette smoke. They then release proteolytic enzymes such as neutrophil elastase and matrix metalloproteinases (MMPs). While these enzymes are meant to destroy harmful pathogens, they also degrade structural proteins like elastin in the alveolar walls, causing irreversible destruction of alveolar tissue, leading to reduced surface area for gas exchange and loss of lung elasticity, characteristics of emphysema. Hence, a higher percentage of segmented neutrophils in the lungs is associated with Emphysema.

```python
import pandas as pd
import numpy as np
from scipy.stats import ttest_ind
import matplotlib.pyplot as plt
import seaborn as sns

# Step 1: Load the CSV files
# Replace the file paths with your actual files
paq_df = pd.read_csv('/content/merged_PAQ.csv')         # contains SEQN and PAQ605
cbc_df = pd.read_csv('/content/merged_CBC.csv')  # contains SEQN and LBDEONO

# Step 2: Merge both datasets on SEQN
df = pd.merge(paq_df, cbc_df, on='SEQN', how='inner')

# Step 3: Clean the data
df_clean = df[(df['PAQ605'].isin([1, 2])) & (~df['LBDEONO'].isna())]

# Step 4: Create groups
group_yes = df_clean[df_clean['PAQ605'] == 1]['LBDEONO']
group_no  = df_clean[df_clean['PAQ605'] == 2]['LBDEONO']

# Step 5: Perform Welchs t-test
t_stat, p_value = ttest_ind(group_yes, group_no, equal_var=False)

# Step 6: Print results
print("T-test comparing LBDEONO by PAQ605 (Physical Activity):")
print("t-statistic:", round(t_stat, 3))
print("p-value:", round(p_value, 3))

# Step 7: Optional Boxplot
sns.boxplot(data=df_clean, x='PAQ605', y='LBDEONO')
plt.xticks([0, 1], ['Yes', 'No'])
plt.xlabel('Vigorous Physical Activity (PAQ605)')
plt.ylabel('Eosinophil Numbers (1000 cells/uL)')
plt.title('Distribution of Eosinophils Number by Physical Activity Status')
plt.show()
```

```python
import pandas as pd
import numpy as np
from scipy.stats import f_oneway
from statsmodels.stats.multicomp import pairwise_tukeyhsd
import matplotlib.pyplot as plt
import seaborn as sns

# Step 1: Load the datasets
sleep_df = pd.read_csv('/content/merged_DPQ.csv')       # contains SEQN and DPQ030
cbc_df = pd.read_csv('/content/merged_CBC.csv')       # contains SEQN and LBDEONO

# Step 2: Merge the datasets
df = pd.merge(sleep_df, cbc_df, on='SEQN', how='inner')

# Step 3: Filter for valid DPQ030 values only
valid_dpq = [1, 2, 3, 4, 5]
df_clean = df[df['DPQ030'].isin(valid_dpq)]

# Step 4: Map category labels
dpq_labels = {
    1: 'Never',
    2: 'Rarely',
    3: 'Sometimes',
    4: 'Often',
    5: 'Always'
}
df_clean['DPQ030_label'] = df_clean['DPQ030'].map(dpq_labels)

# Step 5: Build all groups (including empty ones)
group_dict = {label: df_clean[df_clean['DPQ030_label'] == label]['LBDEONO'] for label in dpq_labels.values()}

# Print number of values per group
print("\nSample sizes per group:")
for label, group in group_dict.items():
    print(f"{label}: n = {group.notna().sum()}")

# Step 6: Filter for groups with at least 2 values (required for ANOVA/Tukey)
valid_groups = [group.dropna() for group in group_dict.values() if group.notna().sum() >= 2]
valid_labels = [label for label, group in group_dict.items() if group.notna().sum() >= 2]

# Step 7: Perform ANOVA (only if 2+ groups are valid)
if len(valid_groups) >= 2:
    f_stat, p_value = f_oneway(*valid_groups)
    print("\nANOVA results")
    print("F-statistic:", round(f_stat, 3))
    print("p-value:", round(p_value, 3))

    # Step 8: Tukey HSD
    data = pd.concat(valid_groups, ignore_index=True)
    group_labels = []
    for label, group in zip(valid_labels, valid_groups):
        group_labels.extend([label] * len(group))

    df_tukey = pd.DataFrame({'value': data, 'group': group_labels})

    tukey = pairwise_tukeyhsd(endog=df_tukey['value'],
                              groups=df_tukey['group'],
                              alpha=0.05)

    print("\nTukey HSD Post-Hoc Test Results:")
    print(tukey)

    # Step 9: Boxplot (all categories, including small ones)
    sns.boxplot(data=df_clean, x='DPQ030_label', y='LBDEONO', order=list(dpq_labels.values()))
    plt.xlabel('Trouble Sleeping Frequency (DPQ030)')
    plt.ylabel('Eosinohils Number (1000 cells/uL)')
    plt.title('Eosinophils Number by Trouble Sleeping Frequency')
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()
else:
    print("Not enough groups with 2 valid values to perform ANOVA.")
```

Since the p-value (0.21) > 0.05, this means that there is no significant difference between trouble sleeping frequency and eosinophils number. Hence, we fail to reject the null hypothesis.

Since the p-value (0.0) < 0.05, this means that there is a significant difference between smoking status and eosinophils number. Hence, we reject the null hypothesis.

## **9.4 Reflection**

####***What did we learn from using AI, and what were its limitations?***
Using AI deepened our understanding of both biomedical research and data analytics. We learned not only how to perform statistical tests and generate visuals in Python (which was not something we were exposed to prior to this module, or even expected to learn during our time here in NP BMS), but also how to interpret what those outputs meant in the context of our research question.

One major insight was that AI does not replace critical thinking, it _enhances_ it.
> For example, while ChatGPT could provide codes to generate visual charts and tables for us, it did not know our specific datasets or what factor/variables we were specifically researching and studying about. This meant we had to manually verify column names, data types and structure before blindly using the generated Python codes we asked for.

Some limitations included:
* AI sometimes hallucinated variable names (e.g., suggesting LBXEONO instead of LBDEONO).
* It occasionally provided code snippets that were incompatible with NHANES-specific survey design, especially when weighting was required.
* Interpretation of statistical results was usually accurate but lacked nuance unless we asked for detailed follow-ups.

We also realized that AI needs precise prompting to work well, as vague questions often led to generic responses that did not help us, whereas well-structured prompts yielded useful, targeted outputs that actually did contribute meaningfully to our research.

####***Reflection on how AI influenced our understanding and approach***
AI reshaped how we approached data analysis by turning what would have been a trial-and-error learning process, into a guided and meaningful experience. It reduced our fear of making continuous hidden mistakes in Python that would take up so much time to find on our own, allowing us to test different analyses efficiently without getting stuck on the initial coding syntax.

More importantly, it encouraged us to connect technical outputs with biomedical meaning.
> For example, after plotting eosinophil trends across NHANES cycles, we were able to ask Gemini, What does this increase in eosinophil prevalence (as indicated by our codes output results for its Q-Q plot and histogram) suggest for respiratory inflammation?

>> This helped us think beyond code and interpret the biological implications of our findings in relation to our target population and hypothesis.

It also encouraged curiosity and thinking of more out-of-the-box questions so that we could provide a more insightful and detailed overall project report. We asked more "what if" questions, and used the answers to drive new analyses we previously never even thought of exploring.

That iterative cycle helped us produce more robust and meaningful results to refine our project to what it is today.

Insights gained:
* AI can be a collaborative research partner, not just a code generator.
* It supports exploratory thinking, especially when dealing with large, messy datasets like NHANES.
* AI helps bridge the gap between statistical methods and their real-world applications in biomedical research.

Challenges encountered:
* Needing to adapt AI-generated code to fit our specific data manually (e.g., renaming columns, handling missing values).
